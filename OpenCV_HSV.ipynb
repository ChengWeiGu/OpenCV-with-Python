```python
#除了区域，图像本身的属性操作也非常多，比如可以通过HSV空间对色调和明暗进行调节。
#HSV空间是由美国的图形学专家A. R. Smith提出的一种颜色空间，HSV分别是色调（Hue），饱和度（Saturation）和明度（Value）。
#在HSV空间中进行调节就避免了直接在RGB空间中调节是还需要考虑三个通道的相关性。OpenCV中H的取值是[0, 180)，其他两个通道的取值都是[0, 256)

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
import cv2
```

```python
path = 'C:\\David\\yellow_tree.jpg'

imag = cv2.imread(path)
print(imag.shape)
plt.imshow(imag[:,:,::-1], interpolation='none')
plt.show()

# Uasually, the HSV space is used to process the image for the color and intensity
imag_HSV = cv2.cvtColor(imag, cv2.COLOR_BGR2HSV)
```

```python
# in H space, we add 15 to each pixel
imag_HSV_green = imag_HSV.copy()
imag_HSV_green[:,:,0] = (imag_HSV_green[:,:,0] + 15) % 180
imag_HSV_green = cv2.cvtColor(imag_HSV_green, cv2.COLOR_HSV2BGR)

plt.imshow(imag_HSV_green[:,:,::-1], interpolation='none')
plt.show()
```

```python
# in S space, we reduce color purity by multiplying 0.5
imag_HSV_color = imag_HSV
imag_HSV_color[:,:,1] = 0.5*imag_HSV_green[:,:,1]
imag_HSV_color = cv2.cvtColor(imag_HSV_color, cv2.COLOR_HSV2BGR)

plt.imshow(imag_HSV_color[:,:,::-1], interpolation='none')
plt.show()
```

```python
# in V space, we multiply values by 0.5 to make the image darker
imag_HSV_dark = imag_HSV.copy()
imag_HSV_dark[:,:,2] = 0.5*imag_HSV_dark[:,:,2]
imag_HSV_dark = cv2.cvtColor(imag_HSV_dark, cv2.COLOR_HSV2BGR)

plt.imshow(imag_HSV_dark[:,:,::-1], interpolation='none')
plt.show()
```

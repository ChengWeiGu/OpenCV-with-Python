```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import cv2
```

```python
path = 'C:\\David\\puppy.jpg'

imag = cv2.imread(path)
print(imag.shape)
plt.imshow(imag[:,:,::-1], interpolation= 'none')
plt.show()
```

```python
# define a M_enlarge matrix to enlarge the imag and do a shift
# note that MX = X' where M is a 2X3 transforming matrix

M_enlarge = np.array([[1.5, 0, -80],
                      [0, 1.5, -50]])

imag_head = cv2.warpAffine(imag, M_enlarge, (300,168))
print(imag_head.shape)
plt.imshow(imag_head[:,:,::-1], interpolation= 'none')
plt.show()
```

```python
# shearing the photos in x direction
 = 30*np.pi/180
M_shear = np.array([[1, np.tan(),0],
                    [0,1,0]])

imag_shear = cv2.warpAffine(imag, M_shear, (300,168))
print(imag_shear.shape)
plt.imshow(imag_shear[:,:,::-1], interpolation= 'none')
plt.show()
```

```python
# to rotate the photo
_rotate = 90*np.pi/180
M_rotate = np.array([[np.cos(_rotate), -np.sin(_rotate), 200],
                     [np.sin(_rotate), np.cos(_rotate), -80]])

imag_rotate = cv2.warpAffine(imag, M_rotate, (300,168))
print(imag_rotate.shape)
plt.imshow(imag_rotate[:,:,::-1], interpolation= 'none')
plt.show()
```

```python
# in general, we can simultaneously do rotation, resize and shifting
M_general = np.array([[1,1.5,-50],
                      [0.5,2,-100]])

imag_general = cv2.warpAffine(imag, M_general, (300,168))
print(imag_general.shape)
plt.imshow(imag_general[:,:,::-1], interpolation= 'none')
plt.show()
```

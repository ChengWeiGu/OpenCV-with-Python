```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
path = 'C:\\David\\dots01.png'

imag = cv2.imread(path, 0)
print(imag.shape)

plt.imshow(imag, 'gray')
plt.show()
```

```python
# define the kernel
kernel = np.ones((15,15))
# erode
imag_erode = cv2.erode(imag, kernel, iterations = 1)
# dilate
imag_dilate = cv2.dilate(imag, kernel, iterations =1)
# eroding after dilating
imag_close = cv2.morphologyEx(imag, cv2.MORPH_CLOSE, kernel)
# dilating after eroding
imag_open = cv2.morphologyEx(imag, cv2.MORPH_OPEN, kernel)

imags = [imag, imag_erode,imag_dilate,imag_close,imag_open]
titles = ['original','erode','dilate','close','open']


plt.figure(figsize=(12,6))
for sub_p , title, im in zip([231,232,233,234,235],titles, imags):
    plt.subplot(sub_p)
    plt.imshow(im, 'gray')
    plt.title(title)
    plt.axis('off')
    
plt.tight_layout(True)
plt.show()
```

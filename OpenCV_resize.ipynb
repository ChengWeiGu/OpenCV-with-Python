```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mp
import cv2
```

```python
path = 'C:\\David\\puppy.jpg'

imag_bgr = cv2.imread(path)
imag_rgb = cv2.cvtColor(imag_bgr, cv2.COLOR_BGR2RGB)
print(imag_bgr.shape)
plt.imshow(imag_rgb, interpolation='none')
plt.show()
```

```python
# resize an image: height X width
imag_200x200 = cv2.resize(imag_bgr,(200,200), interpolation = cv2.INTER_NEAREST)
print(imag_200x200.shape)
plt.imshow(imag_200x200[:,:,::-1], interpolation='none')
plt.show()
```

```python
# resize an image: method2
# if target == (0,0), the size will be curbed by fx and fy
imag_84x150 = cv2.resize(imag_bgr,(0,0), fx = 0.333 , fy = 0.5, interpolation = cv2.INTER_LINEAR)
print(imag_84x150.shape)
plt.imshow(imag_84x150[:,:,::-1], interpolation = 'none')
plt.show()
```

```python
# CopyMakeBorder: adding black pixels
imag_150x250 = cv2.copyMakeBorder(imag_84x150, 16,50,100,50,cv2.BORDER_CONSTANT, value = (0,0,0))
print(imag_150x250.shape)
plt.imshow(imag_150x250[:,:,::-1], interpolation = 'none')
plt.show()
```

```python
# crop the photos
imag_head = imag_bgr[10:150,70:220]
print(imag_head.shape)
plt.imshow(imag_head[:,:,::-1], interpolation = 'none')
plt.show()
```

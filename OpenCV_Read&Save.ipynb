```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
# Read Images in several ways

impath = 'C:\\David\\puppy.jpg'

imag1 = cv2.imread(impath) #BGR chanels
imag2 = cv2.cvtColor(imag1, cv2.COLOR_BGR2RGB)
imag3 = cv2.imread(impath, cv2.IMREAD_GRAYSCALE)
imag4 = mpimg.imread(impath)
im_list = [imag1, imag2, imag3, imag4]

plt.figure(figsize=(14,7))
for i in range(4):
    plt.subplot(2,2,i+1)
    if i == 2:
        plt.imshow(im_list[i], interpolation='none', cmap=plt.cm.gray)
    else:
        plt.imshow(im_list[i], interpolation= 'none')
    plt.axis('off')

plt.show()
```

```python
# Y = 0.299 R + 0.587 G + 0.114 B
for order, im in enumerate(im_list):
    if order+1 == 3:
        print("Image{}[1,1] = {}".format(order+1,im[1,1]))
    else:
        print("Image{}[1,1,:] = {}".format(order+1,im[1,1,:]))

# if we put (R,G,B) into Y formula, we get 180
```
